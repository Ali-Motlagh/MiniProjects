Linear programming is used to solve situations involving deciding the optimal combination of items at scale. 

This program builds a LP for finding the best way to make it up to a partner after you piss them off involving possibly using flowers, massages and/or chocolates as tokens of forgiveness.

The question is; what is the right combination given your wallet and availability that day?

Run the code block and determine your situation, the program will tell you the optimal combination of each item to amend your relationship.

```python
import numpy as np
import gurobipy as gb
```

```python
if __name__ == "__main__":
    x1= int(input(print("Severity of situation [0-100]: ")))
    x2= int(input(print("Maximum budget [16-inf]: ")))
    x3= int(input(print("Time allowance in minutes [4-1440 (24hr)]: ")))
    
    """         Step 1. Sets and indices            """
  

    """         Step 2. Parameters                  """



    """         Step 3. Decision variables          """
    
    model           = gb.Model('Reconcile')
    f     = model.addVar(
                            name     = 'flower',
                            vtype    = gb.GRB.INTEGER,
                            lb       = 0,
                            ub       = gb.GRB.INFINITY )
    m     = model.addVar(
                            name     = 'massage',
                            vtype    = gb.GRB.INTEGER,
                            lb       = 0,
                            ub       =  gb.GRB.INFINITY )
    c     = model.addVar(
                            name     = 'chocolate',
                            vtype    = gb.GRB.INTEGER,
                            lb       = 0,
                            ub       = gb.GRB.INFINITY )


    """         Step 4. Constraints                 """
    model.addConstr(20*f + 35*m + 18*c >= x1) # Value needed to makeup
    model.addConstr(16*f + 32*m +10*c <= x2) # Monetary Cost Budget
    model.addConstr(6*f + 5*m + 7*c <= x3) # Time Allowance Budget



    """         Step 5. Objective function          """
    model.setObjective(20*f + 35*m + 18*c, gb.GRB.MAXIMIZE)

    """         Step 6. Optimize                    """
    print('='*100)
    model.setParam('Method',-1)
    model.update()
    model.optimize()
    print('='*100)

    """         Step 7. Analyze results         """
    print('The optimal flowers:         \t',       f.X)
    print('The optimal massages:         \t',        m.X)
    print('The optimal chocolates:        \t',        c.X)
    print('The optimal value is:    \t',        model.ObjVal)
    print('='*100)
```
